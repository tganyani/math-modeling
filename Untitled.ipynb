```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
g = np.array([9.75,8.375,11.0,9.75,8.5,12.5,9.0,8.5])
w = np.array([27,17,41,26,17,49,23,16])
```

```python
plt.scatter(g,w)
```

```python
y = np.log(w)
x = np.log(g)
```

```python
plt.scatter(x,y)
```

```python
c= np.exp(y.mean() - 3*x.mean())
```

```python
g1 = np.linspace(8,13,20)
w1 = c*g1**3
```

```python
plt.scatter(g,w)
plt.plot(g1,w1,'--')
```
